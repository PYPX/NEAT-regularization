```python
import pickle
```

```python
f = open("iris.data", 'r')

feature = []
target = []
for line in f:
    one_feature = []
    items = line.split(',')
    if(len(items) != 5):
        continue
        
#     print(items)
    for i in range(0,4):
        one_feature.append(float(items[i]))
    feature.append(one_feature)
    target.append(items[-1])
```

```python
set_target = set(target)
list_target = list(set_target)
```

```python
print(list_target)
```

```python
temp = []
for one in target:
    index = list_target.index(one)
    if(index == 0):
        temp.append([1, 0, 0])
    if(index == 1):
        temp.append([0, 1, 0])
    if(index == 2):
        temp.append([0, 0, 1])

target = temp
```

```python
print(target)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
print(scaler.fit(feature))

feature = scaler.transform(feature)
print(feature)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=0)
```

```python
print(len(X_test))
```

```python
with open("train_feature_pickle", 'wb') as f:
    pickle.dump(X_train, f)

with open("train_target_pickle", 'wb') as f:
    pickle.dump(y_train, f)
    
with open("test_feature_pickle", 'wb') as f:
    pickle.dump(X_test, f)
    
with open("test_target_pickle", 'wb') as f:
    pickle.dump(y_test, f)
```
